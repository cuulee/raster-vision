```python
from os.path import join

from skimage import io, transform
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

im_path = join('/opt/data/datasets/planet_kaggle/train-jpg/train_10.jpg')
im = io.imread(im_path)
io.imshow(im)
```

```python
rot_im = transform.rotate(im, 20, mode='constant')
io.imshow(rot_im)
```

```python
rot_im = transform.rotate(im, 20, mode='reflect')
io.imshow(rot_im)
```

```python
# Test that skimage can handle images with > 3 channels
double_im = np.concatenate([im, im], axis=2)
rot_im = transform.rotate(im, 20, mode='reflect')
rot_im = rot_im[:, :, 0:3]
io.imshow(rot_im)
```

```python
import math
scale = 2
tf = transform.AffineTransform(translation=(0, 0), scale=(scale, scale))
warp_im = transform.warp(im, tf, mode='constant')
plt.subplot(1, 2, 1)
plt.imshow(im)
plt.subplot(1, 2, 2)
plt.imshow(warp_im)
```
